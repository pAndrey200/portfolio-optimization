###Yahoo.Finance

Датасет:

* Ticker - биржевой символ
* Type - категория актива
* Date - дата, соответствующая каждому значению в рядах цен; данные за 10 лет с ежедневной частотой
* Open - цена открытия актива; имеются аномальные значения
* High - максимальная цена актива за день
* Low - минимальная цена актива за день; имеются аномальные значения
* Close - цена закрытия актива
* Volume - общее количество единиц актива, проданных за день
* Dividends - выплаченные дивиденды на акцию
* Stock Splits - деление акций; имеются записи до 20
* marketCap - рыночная капитализация
* forwardPE и trailingPE - прогнозный и исторический коэффициенты P/E
* dividendYield - доходность от дивидендов относительно цены

И другие: payoutRatio, priceToBook, bookValue (балансовая стоимость на акцию), debtToEquity, revenue (выручка), grossMargins, operatingMargins, profitMargins (маржинальные коэффициенты на различных этапах доходности), returnOnAssets (ROA), returnOnEquity (ROE), currentRatio и quickRatio (ликвидность).

```python
combined_data = pd.DataFrame()
```

```python
for ticker_symbol in stock_tickers:
    ticker = yf.Ticker(ticker_symbol)
    try:
        historical_data = ticker.history(period=period, interval=interval)
        if historical_data.empty:
            continue
    except Exception as e:
        continue

    df = historical_data.reset_index()
    df.insert(0, 'Ticker', ticker_symbol)
    df.insert(1, 'Type', "stock")

    info = ticker.info
    df['marketCap'] = info.get('marketCap', None)
    df['forwardPE'] = info.get('forwardPE', None)
    df['trailingPE'] = info.get('trailingPE', None)
    df['dividendYield'] = info.get('dividendYield', None)
    df['payoutRatio'] = info.get('payoutRatio', None)
    df['priceToBook'] = info.get('priceToBook', None)
    df['bookValue'] = info.get('bookValue', None)
    df['debtToEquity'] = info.get('debtToEquity', None)
    df['revenue'] = info.get('totalRevenue', None)
    df['grossMargins'] = info.get('grossMargins', None)
    df['operatingMargins'] = info.get('operatingMargins', None)
    df['profitMargins'] = info.get('profitMargins', None)
    df['returnOnAssets'] = info.get('returnOnAssets', None)
    df['returnOnEquity'] = info.get('returnOnEquity', None)
    df['currentRatio'] = info.get('currentRatio', None)
    df['quickRatio'] = info.get('quickRatio', None)

    combined_data = pd.concat([combined_data, df], ignore_index=True)
```

```python
for ticker_symbol in commodity_tickers:
    ticker = yf.Ticker(ticker_symbol)
    try:
        historical_data = ticker.history(period=period, interval=interval)
        if historical_data.empty:
            continue
    except Exception as e:
        continue

    df = historical_data.reset_index()
    df.insert(0, 'Ticker', ticker_symbol)
    df.insert(1, 'Type', "commodity")

    combined_data = pd.concat([combined_data, df], ignore_index=True)
```

```python
for ticker_symbol in bond_tickers:
    ticker = yf.Ticker(ticker_symbol)
    try:
        historical_data = ticker.history(period=period, interval=interval)
        if historical_data.empty:
            continue
    except Exception as e:
        continue

    df = historical_data.reset_index()
    df.insert(0, 'Ticker', ticker_symbol)
    df.insert(1, 'Type', "bond")

    info = ticker.info
    df['marketCap'] = info.get('marketCap', None)
    df['forwardPE'] = info.get('forwardPE', None)
    df['trailingPE'] = info.get('trailingPE', None)
    df['dividendYield'] = info.get('dividendYield', None)
    df['payoutRatio'] = info.get('payoutRatio', None)
    df['priceToBook'] = info.get('priceToBook', None)
    df['bookValue'] = info.get('bookValue', None)
    df['debtToEquity'] = info.get('debtToEquity', None)
    df['revenue'] = info.get('totalRevenue', None)
    df['grossMargins'] = info.get('grossMargins', None)
    df['operatingMargins'] = info.get('operatingMargins', None)
    df['profitMargins'] = info.get('profitMargins', None)
    df['returnOnAssets'] = info.get('returnOnAssets', None)
    df['returnOnEquity'] = info.get('returnOnEquity', None)
    df['currentRatio'] = info.get('currentRatio', None)
    df['quickRatio'] = info.get('quickRatio', None)

    combined_data = pd.concat([combined_data, df], ignore_index=True)


combined_data = combined_data.drop(columns=['Capital Gains'])
```

### EDA - TS

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
csco_data = data[(data['Ticker'] == 'CSCO') & (data['Type'] == 'stock')].copy()
csco_data.set_index('Date', inplace=True)


csco_data.sort_index(inplace=True)
csco_close = csco_data['Close'].dropna()


decomposition = seasonal_decompose(csco_close, model='additive', period=252)
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(10, 8), sharex=True)

ax1.plot(decomposition.observed, label='Observed', color='blue')
ax1.set_ylabel('Observed')
ax1.legend(loc='upper left')

ax2.plot(decomposition.trend, label='Trend', color='orange')
ax2.set_ylabel('Trend')
ax2.legend(loc='upper left')

ax3.plot(decomposition.seasonal, label='Seasonal', color='green')
ax3.set_ylabel('Seasonal')
ax3.legend(loc='upper left')

ax4.plot(decomposition.resid, label='Residual', color='red')
ax4.set_ylabel('Residual')
ax4.legend(loc='upper left')

plt.xlabel('Date')
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
tickers = ['BA', 'VALE', 'CAT', 'OXY', 'GIS', 'T', 'KO', 'AMD', 'CHTR', 'KO']
ticker_data = data[data['Ticker'].isin(tickers)]

adf_results = {}

for ticker in tickers:
    close_prices = ticker_data[ticker_data['Ticker'] == ticker]['Close'].dropna()
    adf_test = adfuller(close_prices)
    adf_results[ticker] = {
        'ADF Stat': adf_test[0],
        'p': adf_test[1],
        'crit val': adf_test[4]
    }

adf_results_df = pd.DataFrame(adf_results).T
adf_results_df
```

Все элементы показывают высокое значение p-value, что говорит о том, что временные ряды цен закрытия для всех данных активов нестационарны. Вероятно, результат для облигаций будет аналогичным.
